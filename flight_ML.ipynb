```python
import numpy as np

def calcular_largura_bin(df, coluna):
    Q75, Q25 = np.percentile(df[coluna], [75 ,25])
    IQR = Q75 - Q25
    largura_bin = 2 * IQR * np.power(len(df[coluna]), -1/3)
    return largura_bin

largura_bin_at = calcular_largura_bin(dados, 'arrival_time')
sns.histplot(data=dados, x='arrival_time', kde=True, binwidth=largura_bin_at)
plt.show()

```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def calcular_metricas_regressao(y_test, y_pred):

    rmse = mean_squared_error(y_test, y_pred, squared=False)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    metricas = {
        'Raiz do Erro Quadrtico Mdio': round(rmse, 4),
        'Erro Absoluto Mdio': round(mae, 4),
        'R2 Score': round(r2, 4)
    }

    return metricas
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dados = pd.read_csv('flights.csv')

# Anlise exploratria
dados.head()
dados.describe()
dados.info()
```

```python
# Transformar variveis categricas
dados['schengen'] = dados['schengen'].replace({'non-schengen': 0, 'schengen': 1})
dados['is_holiday'] = dados['is_holiday'].replace({False: 0, True: 1})

variaveis_categoricas = ['airline', 'aircraft_type', 'origin']
df_encoded = pd.get_dummies(data=dados, columns=variaveis_categoricas, dtype=int)

# Preparar as features e a varivel target
X = df_encoded.drop(['delay', 'flight_id', 'day', 'year'], axis=1)
y = df_encoded['delay']

# Dividir em treino e teste
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score

# Treinar o modelo de Random Forest
model_rf = RandomForestRegressor(random_state=42)
model_rf.fit(X_train, y_train)

# Importncia das features
importances = model_rf.feature_importances_
features = X.columns

feature_importance_df = pd.DataFrame({'feature': features, 'importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='importance', ascending=False)

# Selecionar as top N features mais importantes (ajuste conforme necessrio)
N = 13
selected_features = feature_importance_df['feature'].values[:N]

# Filtrar o conjunto de dados para usar apenas as features selecionadas
X_train_selected = X_train[selected_features]
X_test_selected = X_test[selected_features]
```

```python
# Treinar o modelo de regresso linear com as features selecionadas
from sklearn.linear_model import LinearRegression

model_lr = LinearRegression()
model_lr.fit(X_train_selected, y_train)

# Fazer previses e avaliar o modelo
y_pred_lr = model_lr.predict(X_test_selected)
mae_lr = mean_absolute_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)

print(f'Mean Absolute Error do modelo de Regresso Linear: {mae_lr}')
print(f'R^2 do modelo de Regresso Linear: {r2_lr}')
```
